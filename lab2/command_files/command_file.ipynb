## UporzÄ…dkowanie zbioru danych - billboard.csv

```python
import numpy as np
import pandas as pd
from itables import show
import itables.options as opt
opt.maxBytes = 0
opt.maxColumns = 100
```

### Surowe dane

```python
data = pd.read_csv(r'..\original_data\billboard.csv', encoding = "ISO-8859-1")
show(data)
```

### Dane po operacji melt

```python
data_melt = data.melt(id_vars=['year', 'artist.inverted', 'track', 'time', 'genre', 'date.entered', 'date.peaked'], var_name='week', value_name='position')
show(data_melt.head(100))
```

### Dane po usuniÄ™ciu wartoÅ›ci NaN

```python
data_clean = data_melt[pd.notna(data_melt['position'])]
show(data_clean.head(500))
data_clean.to_csv(r'..\analysis_data\billboard_clean.csv')
```
