Zadanie 3:
Definicja funkcji oraz wyrysowanie wykres√≥w

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def f(x):
    return x**2 + 5
```

```python
# 1. 
x = np.linspace(-1, 1)
y = f(x)
plt.figure()
plt.grid()
plt.title('x^2 + 5')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
```

```python
# 2. 
x = np.linspace(-6, 6)
y = f(x)
plt.figure()
plt.grid()
plt.title('x^2 + 5')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
```
